```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,1,100)
```

```python
def sin_function(x):
    return np.sin(x)
def cos_function(x):
    return np.cos(x)
    
```

```python
y_sin = sin_function(x)
y_cos = cos_function(x)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10,4))

ax1.plot(x, y_sin)
ax1.set_title("sin_(x)")
ax1.set_xlabel("x")
ax1.set_ylabel("y")

ax2.plot(x, y_cos)
ax2.set_title("cos(x)")
ax2.set_xlabel("x")
ax2.set_ylabel("y")

plt.savefig("session_8_prompt.pdf")
plt.show()
```
